```python
from Bio import SeqIO
import matplotlib.pyplot as plt
```

```python
seq_a = SeqIO.read("KY352407.fasta", "fasta").seq
seq_b = SeqIO.parse("OQ297732.fasta", "fasta")

print("Sequence A")
print(seq_a)

print()

print("Sequence B")
for record in seq_b:
    seq_b = record.seq

print(seq_b)
```

```python
ambiguous_letters = ["N", "Y", "R", "W", "M", "S", "K"]

seq_a = seq_a.replace("N", "")
seq_a = seq_a.replace("Y", "")
seq_a = seq_a.replace("R", "")
seq_a = seq_a.replace("W", "")
seq_a = seq_a.replace("M", "")
seq_a = seq_a.replace("S", "")
seq_a = seq_a.replace("K", "")

seq_b = seq_b.replace("N", "")
seq_b = seq_b.replace("Y", "")
seq_b = seq_b.replace("R", "")
seq_b = seq_b.replace("W", "")
seq_b = seq_b.replace("M", "")
seq_b = seq_b.replace("S", "")
seq_b = seq_b.replace("K", "")
```

```python
from collections import Counter

print(f"Sequence A: {seq_a}")
print(f"Sequence B: {seq_b}")

Counter(seq_a)
```

```python
# Find and display the total length of Sequence A and B
print(len(seq_a))
print(len(seq_b))
```

```python
# Find and display which index the first time the codon (CGC) appears in Sequence A and B.
find = 'CGC'
print(f"First index CGC Sequence A: {seq_a.find(find)}")
print(f"First index CGC Sequence B: {seq_b.find(find)}")
```

```python
# Take the first 6 nucleotides from Sequence A and the last 9 nucleotides from Sequence B then combine the sequences and reverse it to a new sequence, Sequence C

split_seq_a = seq_a[:6]
split_seq_b = seq_b[-9:]

seq_c = (split_seq_a + split_seq_b)[::-1]
seq_c
```

```python
# Determine and display the melting temperature for Sequence A, B, and C (using built-in Wallace Method)
from Bio.SeqUtils import MeltingTemp as mt

mt_a = mt.Tm_Wallace(seq_a)
mt_b = mt.Tm_Wallace(seq_b)
mt_c = mt.Tm_Wallace(seq_c)

print(f"Melting point Seq A: {mt_a}")
print(f"Melting point Seq B: {mt_b}")
print(f"Melting point Seq C: {mt_c}")
```

```python
# Determine and display the GC & AT content percentage for Sequence A, B, and C
from Bio.SeqUtils import GC

gc_a = GC(seq_a)
gc_b = GC(seq_b)
gc_c = GC(seq_c)

at_a = 100 - gc_a
at_b = 100 - gc_b
at_c = 100 - gc_c

print(f"GC A: {gc_a}% AT A: {at_a}%")
print(f"GC B: {gc_b}% AT B: {at_b}%")
print(f"GC C: {gc_c}% AT C: {at_c}%")
```

```python
# Determine and display the molecular weight for Sequence A, B, and C
from Bio.SeqUtils import molecular_weight as mw

mw_a = mw(seq_a)
mw_b = mw(seq_b)
mw_c = mw(seq_c)

print(f"MW A: {gc_a}")
print(f"MW B: {gc_b}")
print(f"MW C: {gc_c}")
```

```python
# Determine and display the nucleotides base (A, C, G, T) frequency for Sequence A, B, and C with a chart usmg matplotlib library
from collections import Counter

freq_a = Counter(seq_a)
freq_b = Counter(seq_b)
freq_c = Counter(seq_c)

print(f"Frequency A: {freq_a}")
print(f"Frequency B: {freq_b}")
print(f"Frequency C: {freq_c}")
```

```python
plt.figure(figsize=(12, 12)) # Ganti ukuran plotting

# subplot(row, column, slot)
plt.subplot(1, 3, 1)
plt.title("Base Frequency A")
plt.pie(freq_a.values(), labels=freq_a.keys())

plt.subplot(1, 3, 2)
plt.title("Base Frequency B")
plt.pie(freq_b.values(), labels=freq_b.keys())

plt.subplot(1, 3, 3)
plt.title("Base Frequency C")
plt.pie(freq_c.values(), labels=freq_c.keys())
plt.show()
```

```python
"""
Transcribe and display the DNA Sequence A, B: and C into mRNA Sequences.
Translate and display the mRNA Sequence A, B, and C Into amino acids sequences-
"""

mrna_a = seq_a.transcribe()
mrna_b = seq_b.transcribe()
mrna_c = seq_c.transcribe()

print(f"mRNA A: {mrna_a}")
print(f"mRNA B: {mrna_b}")
print(f"mRNA C: {mrna_c}")

amino_a = mrna_a.translate()
amino_b = mrna_b.translate()
amino_c = mrna_c.translate()
print()
print(f"Amino Acid mRNA A: {amino_a}")
print(f"Amino Acid mRNA B: {amino_b}")
print(f"Amino Acid mRNA C: {amino_c}")
```

```python
# Perform local alignment pairwise between Sequence A and Sequence "TGTACATTATAATGCTCTATTTTCTCGC" and display the alignments' score.
from Bio.Seq import Seq
from Bio import pairwise2

seq_x = Seq("TGTACATTATAATGCTCTATTTTCTCGC")

# local_alignment_a_x = pairwise2.align.localxx(seq_a[:50], seq_x[:50])
# for alignment in local_alignment_a_x:
#     print(pairwise2.format_alignment(*alignment))

local_alignment_a_x_score = pairwise2.align.localxx(
    seq_a, seq_x, one_alignment_only=True, score_only=True
)

local_alignment_a_x_score
```

```python
# dotplot


def dot_plot(a, b):
    fill = chr(0x25E9)
    blank = " "
    print(" |" + a)
    for i in b:
        print(i + "|", end="")
        for j in a:
            if i == j:
                print(fill, end="")
            else:
                print(blank, end="")
        print()

dot_plot(seq_a[:15], seq_b[:15])
```

```python
import numpy as np

def delta(a, b):
    return 0 if a == b else 1

def M(seq1, seq2, i, j, k):
    return sum(delta(x, y) for x, y in zip(seq1[i:i+k], seq2[j:j+k]))

def makeMatrix(seq1, seq2, k):
    n = len(seq1)
    m = len(seq2)
    return [[M(seq1, seq2, i, j, k) for j in range(m-k+1)] for i in range(n-k+1)]

def dotplot(seqA, seqB):
    plt.imshow(np.array(makeMatrix(seqA, seqB, 1)))
    plt.xticks(np.arange(len(seqA)), seqA)
    plt.yticks(np.arange(len(seqB)), seqB)
    plt.show()

dotplot(seq_a[:15], seq_b[:15])
```
