```python
from Bio import SeqIO

seqA = (SeqIO.read("KY352407.fasta", "fasta"))
seqB = (SeqIO.read("OQ297732.fasta", "fasta"))

sequenceA = (seqA.seq)
sequenceB = (seqB.seq)
sequenceA = sequenceA.replace("N", "")
sequenceA = sequenceA.replace("Y", "")
sequenceA = sequenceA.replace("R", "")
sequenceA = sequenceA.replace("W", "")
sequenceA = sequenceA.replace("M", "")
sequenceA = sequenceA.replace("S", "")
sequenceA = sequenceA.replace("K", "")

sequenceB = sequenceB.replace("N", "")
sequenceB = sequenceB.replace("Y", "")
sequenceB = sequenceB.replace("R", "")
sequenceB = sequenceB.replace("W", "")
sequenceB = sequenceB.replace("M", "")
sequenceB = sequenceB.replace("S", "")
sequenceB = sequenceB.replace("K", "")

print(sequenceA)
print(sequenceB)
```

```python
# Sequence Manipulation
# a. Find and display the total length of Sequence A and B
seqAlength = len(sequenceA)
seqBlength = len(sequenceB)
print("Sequence A length: ", seqAlength)
print("Sequence B length: ", seqBlength)
```

```python
# b. Find and display which index the first time the codon (CGC) appears in Sequence A and B
print(str(sequenceA).index("CGC"))
print(str(sequenceB).index("CGC"))
```

```python
# c. Take the first 6 nucleotides from Sequence A and the last 9 nucleotides from Sequence B then combine the sequences and reverse it to a new sequence, Sequence C
sequenceC = sequenceA[:6] + sequenceB[-9:]
sequenceC = sequenceC[::-1]
print(sequenceC)
```

```python
# Sequence Analysis & Plotting
from matplotlib import pyplot as plt
from Bio.SeqUtils import MeltingTemp as mt
# a. Determine and display the melting temperature for Sequence A, B, and C (using built-in Wallace Method)
meltA = mt.Tm_Wallace(sequenceA)
meltB = mt.Tm_Wallace(sequenceB)
print("Melting temperature A =", meltA)
print("Melting temperature B =", meltB)
```

```python
# b. Determine and display the GC & AT content percentage for Sequence A, B, and C
from Bio.SeqUtils import GC
gcA = GC(sequenceA)
gcB = GC(sequenceB)
gcC = GC(sequenceC)
print("GC of sequence A =", gcA)
print("GC of sequence B =", gcB)
print("GC of sequence C =", gcC)

atA = 100 - gcA
atB = 100 - gcB
atC = 100 - gcC
print("AT of sequence A =", atA)
print("AT of sequence B =", atB)
print("AT of sequence C =", atC)
```

```python
# c. Determine and display the molecular weight for Sequence A, B, and C
from Bio.SeqUtils import molecular_weight as mw
mwA = mw(sequenceA)
mwB = mw(sequenceB)
mwC = mw(sequenceC)
print("Molecular weight of sequence A =", mwA)
print("Molecular weight of sequence B =", mwB)
print("Molecular weight of sequence C =", mwC)
```

```python
# d. Determine and display the nucleotides base (A, C, G, T) frequency for Sequence A, B, and C with a chart using matplotlib library
from collections import Counter
countA = Counter(sequenceA)
countB = Counter(sequenceB)
countC = Counter(sequenceC)
plt.bar(countA.keys(), countA.values())
plt.title("Sequence A")
plt.show()
plt.bar(countB.keys(), countB.values())
plt.title("Sequence B")
plt.show()
plt.bar(countC.keys(), countC.values())
plt.title("Sequence C")
plt.show()
```

```python
# DNA & mRNA Protein Synthesis
# a. Transcribe and display the DNA Sequence A, B, and C into mRNA Sequences
print("Transcribe A =", sequenceA.transcribe())
print("Transcribe B =", sequenceB.transcribe())
print("Transcribe C =", sequenceC.transcribe())
```

```python
# b. Translate and display the mRNA Sequences A, B, and C into amino acids sequences
print("Translate A =", sequenceA.translate())
print("Translate B =", sequenceB.translate())
print("Translate C =", sequenceC.translate())
```

```python
# Sequence Alignment & Similarities
# a. Performn local alignment pairwise between Sequence A and Sequence "TGTACATTATAATGCTCTATTTTCTCGC" and display the alignments' score.
from Bio.Seq import Seq
from Bio import pairwise2
seqNew = Seq('TGTACATTATAATGCTCTATTTTCTCGC')
print(pairwise2.align.localxx(sequenceA, seqNew, one_alignment_only=True, score_only=True))
```

```python
# b. Performn local alignment pairwise between Sequence B and Sequence "CAGAGTAATGTGTACAGATATTGGGCTC" and display the alignments' score.
seqX = Seq("CAGAGTAATGTGTACAGATATTGGGCTC")
print(pairwise2.align.localxx(sequenceA, seqX, one_alignment_only=True, score_only=True))
```
