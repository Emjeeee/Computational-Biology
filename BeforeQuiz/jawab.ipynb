```python
from Bio import SeqIO
# seqA = SeqIO.read("KY352407.fasta", "fasta").seq
# seqB = SeqIO.read("OQ297732.fasta", "fasta").seq

for x in SeqIO.parse("KY352407.fasta", "fasta"):
    seqA = x.seq

for x in SeqIO.parse("OQ297732.fasta", "fasta"):
    seqB = x.seq

print(seqA)
print(seqB)
```

```python
# remove ambiguous letters
seqA = seqA.replace('N', '')
seqA = seqA.replace('Y', '')
seqA = seqA.replace('R', '')
seqA = seqA.replace('W', '')
seqA = seqA.replace('M', '')
seqA = seqA.replace('S', '')
seqA = seqA.replace('K', '')

seqB = seqB.replace('N', '')
seqB = seqB.replace('Y', '')
seqB = seqB.replace('R', '')
seqB = seqB.replace('W', '')
seqB = seqB.replace('M', '')
seqB = seqB.replace('S', '')
seqB = seqB.replace('K', '')

print(seqA)
print(seqB)
```

```python
# a.
print("Total length =", len(seqA))
print("Total length =", len(seqB))
```

```python
# b.
print("Index =", seqA.find('CGC'))
print("Index =", seqB.find('CGC'))
```

```python
# c.
seqC = seqA[:6] + seqB[-9:]
seqC = seqC[::-1]
print("Sequence C =", seqC)
```

```python
from Bio.SeqUtils import MeltingTemp as mt
print("Melting Temp seq A =", mt.Tm_Wallace(seqA))
print("Melting Temp seq B =", mt.Tm_Wallace(seqB))
print("Melting Temp seq C =", mt.Tm_Wallace(seqC))
```

```python
from Bio.SeqUtils import GC
print("GC seq A =", GC(seqA))
print("AT seq A =", 100 - GC(seqA))
print()
print("GC seq B =", GC(seqB))
print("AT seq B =", 100 - GC(seqB))
print()
print("GC seq C =", GC(seqC))
print("AT seq C =", 100 - GC(seqC))
```

```python
from Bio.SeqUtils import molecular_weight as mw
print("Molecular Weight seq A =", mw(seqA))
print("Molecular Weight seq B =", mw(seqB))
print("Molecular Weight seq C =", mw(seqC))
```

```python
from collections import Counter
counterA = Counter(seqA)
counterB = Counter(seqB)
counterC = Counter(seqC)

from matplotlib import pyplot as plt
plt.bar(counterA.keys(), counterA.values())
plt.title("Sequence A")
plt.show()
plt.bar(counterB.keys(), counterB.values())
plt.title("Sequence B")
plt.show()
plt.bar(counterC.keys(), counterC.values())
plt.title("Sequence C")
plt.show()

plt.figure(figsize=(12, 12))

plt.subplot(1, 3, 1)
plt.title("Sequence A")
plt.pie(counterA.values(), labels=counterA.keys())

plt.subplot(1, 3, 2)
plt.title("Sequence B")
plt.pie(counterB.values(), labels=counterB.keys())

plt.subplot(1, 3, 3)
plt.title("Sequence C")
plt.pie(counterC.values(), labels=counterC.keys())
plt.show()
```

```python
mrna_a = seqA.transcribe()
mrna_b = seqB.transcribe()
mrna_c = seqC.transcribe()
print(mrna_a)
print(mrna_b)
print(mrna_c)
print()
print("mRNA seq A =", seqA.transcribe())
print("mRNA seq B =", seqB.transcribe())
print("mRNA seq C =", seqC.transcribe())

print("\n")

print("amino acids seq A =", mrna_a.translate())
print("amino acids seq B =", mrna_b.translate())
print("amino acids seq C =", mrna_c.translate())
print()
print("amino acids seq A =", seqA.translate())
print("amino acids seq B =", seqB.translate())
print("amino acids seq C =", seqC.translate())
```

```python
from Bio import pairwise2
from Bio.Seq import Seq
seqX = Seq('TGTACATTATAATGCTCTATTTTCTCGC')
print("Local alignment seq A & seq X =", pairwise2.align.localxx(seqA, seqX, one_alignment_only=True, score_only=True))

seqY = Seq('CAGAGTAATGTGTACAGATATTGGGCTC')
print("Local alignment seq B & seq Y =", pairwise2.align.localxx(seqB, seqY, one_alignment_only=True, score_only=True))
```

```python
from Levenshtein import distance
print("Distance Levenshtein =", distance(seqA, seqB))
```

```python
def hamming(seqA, seqB):
    return len(
        [
            (x, y) for x, y in zip(seqA, seqB) if x != y
        ]
    )
print("Distance Hamming =", hamming(seqA, seqB))
```

```python
# dotplot
def dotPlot(a, b):
    fill = chr(0x25CF)
    blank = " "
    print(" |" + a)
    for i in b:
        print(i + "|", end="")
        for j in a:
            if i == j:
                print(fill, end="")
            else:
                print(blank, end="")
        print()

dotPlot(seqA[:10], seqB[:10])
```
