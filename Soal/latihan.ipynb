```python
from Bio import SeqIO

# Dari file fasta dan genbank yang ada print can id, description dan seqnya

# cara parse
print("Cara Parse")
print("")
for x in SeqIO.parse("sequence.fasta", "fasta"):
    print("Id =", x.id)
    print("Description =", x.description)
    print("Sequence =", x.seq)
    # fasta
    sequenceFasta = x.seq

print("")

for x in SeqIO.parse("sequence.gb", "genbank"):
    print("Id =", x.id)
    print("Description =", x.description)
    print("Sequence =", x.seq)

print("\n")

# cara read
print("Cara Read")
print("")
fasta = SeqIO.read("sequence.fasta", "fasta")
print("Id =", fasta.id)
print("Description =", fasta.description)
print("Sequence =",fasta.seq)

print("")

genbank = SeqIO.read("sequence.gb", "genbank")
print("Id =", genbank.id)
print("Description =", genbank.description)
print("Sequence =", genbank.seq)

# gen
sequenceGenbank = genbank.seq
```

```python
seq1 = str(sequenceFasta[::-1])
seq2 = str(sequenceFasta)
if(seq1 == seq2):
    print("The sequence is palindrome")
else:
    print("Not")

def isPalindrome(seq):
    seq = seq.lower()
    length = len(seq)
    for i in range(length // 2):
        if seq[i] != seq[length - i -1]:
            return False
    return True

if isPalindrome(seq2):
    print("Palindrome")
else:
    print("Not palindrome")
```
